```python

import numpy as np
def get(path):
    times = []
    values = []
    with open(path, 'r') as inp:
        for line in inp:
            line = line.replace('[', '')
            line = line.replace(']', '')
            line = line.replace(',', '')
            #print(line)
            parts = list(map(float, line.split()))
            times.append(parts[0])
            #values.append(parts[1])
            values.append(np.mean(parts[1:]))
    return times, values
```

```python
def single_plot(path):
    t1, res1 = get(path)
    plt.plot(t1, res1, 'b', label = 'single plot')
```

```python
single_plot('DoomDefendLine-v0/DoomDefendLine-v0')
```

```python
path1 = 'DoomDefendCenter-v0/DoomDefendCenter-v0'
path2 = 'final/default_amn'
t1, res1 = get(path1)
t2, res2 = get(path2)
t2 = t2[:len(t1)]
res2 = res2[:len(res1)]
assert t1 == t2
        
    
```
